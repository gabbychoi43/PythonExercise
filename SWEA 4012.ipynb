```python

import itertools

def Partition(N) :
    X=set(range(N))
    s=it.combinations(X,N//2)
    a=[]
    for i in s :
        a.append([list(set(i)),list(X-set(i))])

    return a

T = int(input())
# 여러개의 테스트 케이스가 주어지므로, 각각을 처리합니다.
for test_case in range(1, T + 1):
    
    N=int(input())
    Material=[]
    Synergy=[[0 for i in range(N)] for i in range(N) ]    
    for i in range(N) :
        tempM=[int(i) for i in input().split()]  
        Material.append(tempM)
    sum=0
    Prob=Partition(N)
    min=2e9
    
    for i in Prob :
        
        for j in i[0] :
            for k in i[0] :
                sum+=Material[j][k]
                
        for j in i[1] :
            for k in i[1] :
                sum-=Material[j][k]
        if abs(sum)<min :
            min=abs(sum)
    print(Material)
    print('#{}'.format(test_case),min)
```
