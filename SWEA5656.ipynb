```python
import itertools as it
import copy

'''

최종제출 코드
'''

def explosion(Loc,m) :  # Blocknum 블록에 써진 숫자, Loc 위치, map은 지도
    Blocknum=m[Loc[0]][Loc[1]]
    m[Loc[0]][Loc[1]]=0
    if Blocknum>1 :
        for i in range(Blocknum) :
            explosion([Loc[0],max(0,Loc[1]-i)],m)
            explosion([Loc[0],min(len(m[0])-1,Loc[1]+i)],m)
            explosion([max(0,Loc[0]-i),Loc[1]],m)
            explosion([min(len(m)-1,Loc[0]+i),Loc[1]],m)
    else :
        return m

def drop(m) :
    droped=copy.deepcopy(m)
    K=[]
    for i in range(len(m)) :
        count=0
        while 0 in droped[i] :
            droped[i].remove(0)
            count+=1
        l=[0]*count+droped[i]
        K.append(l)
    m=K   
    return m           

def countblock(m) :
    count=0
    for i in m :
        for j in i :
            if j!=0 :
                count+=1
    return count

T= int(input())

for test_case in range(1,T+1) :
    L=[int(i) for i in input().split()]
    N=L[0]
    col=L[1]
    row=L[2]
    Map=[]
    count=[]
    for i in range(row) :
        Map.append([int(i) for i in input().split()])
	
    TransMap=list(map(list,(zip(*Map))))
    EMap=copy.deepcopy(TransMap)
    
    
    Prob=list(map(list,it.product(range(col),repeat=N)))
    
    for i in Prob :
        EMap=copy.deepcopy(TransMap)
        for j in i :
            for k in range(row) :
                if EMap[j][k]!=0 :
                    explosion([j,k],EMap)
                    EMap=drop(EMap)
                    break
        count.append(countblock(EMap))

    print('#{}'.format(test_case),min(count))
	
        
```
